```python
import pandas as pd
import matplotlib.image as mpimg
import os
import matplotlib.lines as mlines
#def main():
    
# File paths
parent_dir = '/mnt/data10/shared/yujie/raffat_data/Raw Data/NA/A2/SM/230616Spheroids-4-A2_left_NA_rawDATA_inner'
file_intensity = os.path.join(parent_dir,'inner region-Total Ion Count_col.csv')
file_coordinates = os.path.join(parent_dir,'230616Spheroids-4-A2_left_NA_rawDATA_regionSpots.csv')
feature_list_path = os.path.join(parent_dir,'04-PeakList-PyMTSphingolipidProfile-Pos-2.csv')

skip_rows_intensity = 10  # This number may need to be adjusted based on the actual file content
df_intensity = pd.read_csv(file_intensity, skiprows=skip_rows_intensity, header=0, delimiter=';')

df_feature_list = pd.read_csv(feature_list_path, delimiter=';', skiprows=8)

def find_closest_mz_name(mz_value, df_feature_list):
    # Find the closest m/z value
    closest_match = df_feature_list.iloc[(df_feature_list['m/z'] - mz_value).abs().argsort()[:1]]
    return closest_match['Name'].values[0]
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


# Separating out the features (intensity data)
X = df_intensity.iloc[:, 1:]  # Assuming first column is 'm/z' and should be excluded

# Standardizing the features
X_standardized = StandardScaler().fit_transform(X)

# Performing PCA
pca = PCA(n_components=8)
principal_components = pca.fit_transform(X_standardized)

# Creating a DataFrame with the principal components
pca_df = pd.DataFrame(data=principal_components,
                      columns=['PC1', 'PC2', 'PC3', 'PC4', 'PC5', 'PC6', 'PC7', 'PC8'])

# Concatenating the molecule names ('m/z') with the principal components
final_df = pd.concat([df_intensity['m/z'], pca_df], axis=1)

# Displaying the results
print(final_df)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

def plot_bar(final_df,df_feature_list,col='PC1',num=10):

    # Sort by PC1 and select the top and bottom 10
    top_10 = final_df.nlargest(num, 'PC1')
    bottom_10 = final_df.nsmallest(num, 'PC1')

    # Combine top and bottom into one DataFrame and sort
    selected_molecules = pd.concat([top_10, bottom_10]).sort_values(by='PC1')

    # Replace m/z with mz_name
    selected_molecules['mz_name'] = [find_closest_mz_name(mz, df_feature_list) for mz in selected_molecules['m/z']]

    # Plotting
    plt.figure(figsize=(12, 6))
    x_positions = range(1, len(selected_molecules) + 1)

    for x, row in zip(x_positions, selected_molecules.itertuples()):
        color = 'red' if row.PC1 > 0 else 'blue'
        plt.bar(x, row.PC1, color=color)

    plt.xticks(x_positions, selected_molecules['mz_name'], rotation=45, ha='right')
    plt.gca().spines['bottom'].set_visible(False)

    plt.xlabel('Molecule Name')
    plt.ylabel('PC1 Value')
    plt.title('Top and Bottom 10 Molecules on PC1')
    plt.axhline(0, color='black', linewidth=0.8)
    plt.tight_layout()
    plt.show()
```

```python
plot_bar(final_df,df_feature_list,col='PC1',num=10)
```

```python
plot_bar(final_df,df_feature_list,col='PC2',num=10)
```
